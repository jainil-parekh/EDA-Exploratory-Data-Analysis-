```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
houseprice = pd.read_csv('Bengaluru_House_Data.csv')
```

```python
houseprice.head()
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

1. Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', style='smoker', size='size')
```

2. Bar Plot (Numerical - Categorical)

```python
houseprice.head()
```

```python
sns.barplot(x='area_type', y='price', data=houseprice)
```

3. HeatMap

```python
sns.heatmap(pd.crosstab(houseprice['balcony'] ,houseprice['area_type']))
```

```python
sns.barplot(houseprice.groupby('area_type').mean()['balcony'])
```

4. Cluster Map

```python
sns.clustermap(pd.crosstab(houseprice['balcony'] ,houseprice['area_type']))
```

5. Pairplot - Collection of Scatter Plot

```python
iris.head()
```

```python
sns.pairplot(iris, hue='species')
```

7. Lineplot - We generally use this when we have date-time based (i.e. year, month, day etc) on x-axis

```python
flights.head()
```

```python
new = flights.groupby('year').sum().reset_index()
```

```python
sns.lineplot(x='year', y='passengers', data=new)
```

```python
sns.heatmap(flights.pivot_table(values='passengers', index='month',columns='year'))
```

```python
sns.lineplot(flights.pivot_table(values='passengers', index='month',columns='year'))
```
